```python
print("Data Quality Summary: \n")

# Null value:
null_count = df.isna().sum()

# Empty-like values (empty string or false string)
empty_like = ["na", "n/a", "nan", "none", "unknown", "-", ""]
empty_like_count = df.apply(lambda col: col.astype(str).str.strip().str.lower().isin(empty_like).sum())

# Combine together
summary = pd.DataFrame({
    "null count": null_count,
    "Empty value": empty_like_count
}).fillna(0)

# Total Missing:
summary["Total Missing"] = summary["null count"] + summary["Empty value"]

# Percentage
summary["Missing %"] = (summary["Total Missing"] / len(df) * 100)

# Problematic rows and columns
summary = summary[summary["Total Missing"] > 0]

# Unique value
summary["Unique value"] = df.nunique()

# Data type
summary["Data Type"] = df.dtypes

# Duplicate value
dup_val = df.duplicated().sum()

# Sort
summary = summary.sort_values(by="Total Missing", ascending=False)

print(summary)
print(f"Duplicate Values: {dup_val}")
```

## Data cleaning

```python
# Define the anomaly placeholders we found
placeholders = [-1, '-1', 'na', 'Unknown / Non-Applicable', '']
total_rows = len(df)

# Calculate percentage of errors/missing values BEFORE cleaning
before_missing_pct = ((df.isin(placeholders).sum() + df.isna().sum()) / total_rows) * 100
```

```python
# Handle Missing values 

# Replace variations of missing data with np.nan
missing_indicators = [-1, '-1', 'na', 'Unknown / Non-Applicable']
df.replace(missing_indicators, np.nan, inplace=True)
```

```python
# droping unnecessary coloums 
# 1. List of extra feature columns to drop
columns_to_drop = [
    'my_comments', 'my_tags', 'my_sns', 'my_discuss', 
    'my_rewatch_value', 'my_rewatching', 'my_rewatching_ep', 
    'update_on_import', 'my_priority', 'my_times_watched', 
    'my_storage','my_storage_value','my_id', 'my_rated'
]

# 2. Automatically find all columns that start with "user_"
user_columns = [col for col in df.columns if col.startswith('user_')]

# 3. Combine both lists and drop them 
# (errors='ignore' prevents errors if a column is already gone)
df.drop(columns=columns_to_drop + user_columns, inplace=True, errors='ignore')

# 4. CRITICAL: Drop the very first row (Index 0) because it only contained user stats
df = df.iloc[1:].reset_index(drop=True)

# Look at your beautifully clean dataset!
display(df.head())
```

```python
# Save Cleand dataset
df.info()
df.isnull().sum()

# save to csv file 
df.to_csv("myanimelist_cleaned.csv", index=False)
```

## Data Quality graph after cleaning

```python
# Calculate percentage of missing values AFTER cleaning
after_missing_pct = (df.isna().sum() / total_rows) * 100

# Combine into a single DataFrame for easy plotting
comparison_df = pd.DataFrame({
    'Before Cleaning (%)': before_missing_pct,
    'After Cleaning (%)': after_missing_pct
})

# Keep only columns that actually had errors so the chart isn't cluttered
comparison_df = comparison_df[comparison_df['Before Cleaning (%)'] > 0].sort_values(by='Before Cleaning (%)', ascending=False)

# Plot the Before vs After comparison
plt.figure(figsize=(12, 6))
comparison_df.plot(kind='bar', figsize=(14, 6), color=['#d9534f', '#5cb85c'], width=0.8)

plt.title('Data Quality Check: Anomalies & Missing Values (Before vs. After Cleaning)', fontsize=14)
plt.ylabel('Percentage of Data (%)', fontsize=12)
plt.xlabel('Dataset Columns', fontsize=12)
plt.xticks(rotation=45, ha='right')
plt.legend(loc='upper right')
plt.tight_layout()
plt.show()
```

```python
# My Custome Dashboard


# ---------------------------------------------------------
# CUSTOM ANIME THEME & COLOR PALETTE
# ---------------------------------------------------------
# Set a custom dark theme (Eminence in Shadow vibe)
sns.set_theme(style="darkgrid", rc={
    "figure.facecolor": "#121212",  # Deep dark background
    "axes.facecolor": "#1E1E2E",    # Slightly lighter dark slate for plot backgrounds
    "text.color": "white",
    "axes.labelcolor": "white",
    "xtick.color": "white",
    "ytick.color": "white",
    "grid.color": "#333344",
    "axes.edgecolor": "#555566"
})

# Custom Hex Palette: DBZ Orange, SSJ Gold, One Piece Blue, Luffy Red, Shadow Purple, Slime Cyan
anime_colors = ['#FF6F00', "#45116A", '#005A9C', '#DC143C', '#8A2BE2', '#00FFFF', '#FF4500', '#1E90FF']
sns.set_palette(sns.color_palette(anime_colors))
anime_cmap = sns.color_palette(anime_colors)

# Calculate KPIs
total_watch_time = df['my_watched_episodes'].sum()
avg_episode_length = 24  # in minutes
total_watch_time_hours = total_watch_time * avg_episode_length / 60
highest_rating = df['my_score'].max()
top_rated_anime = df.loc[df['my_score'] == highest_rating, 'series_title'].values[0]

# Set up the figure and axes
fig = plt.figure(figsize=(22, 20))
fig.suptitle('Ultimate Anime Tracking Dashboard', fontsize=28, fontweight='bold', y=0.97, color='#FFD700') # Gold Title

# Add KPIs as text at the top
kpi_text = (
    f"Total Watch Time: {total_watch_time_hours:.0f} hours\n"
    f"Highest Rating: {highest_rating} (Anime: {top_rated_anime})"
)
# Updated bbox for dark mode
fig.text(0.5, 0.91, kpi_text, ha='center', fontsize=16, fontweight='bold',
         bbox=dict(facecolor='#1E1E2E', edgecolor='#8A2BE2', alpha=0.9, boxstyle='round,pad=0.8', linewidth=2))

# Create a grid for the plots
gs = fig.add_gridspec(3, 3)

# ---------------------------------------------------------
# Plot 1: Distribution of Anime Scores
ax1 = fig.add_subplot(gs[0, 0])
sns.histplot(df['my_score'], bins=10, kde=True, color='#00FFFF', ax=ax1, edgecolor='#121212', alpha=0.7)
ax1.set_title('Distribution of Anime Scores', fontsize=15, pad=15, fontweight='bold')

# Plot 2: Watching Progress Over Time 
ax2 = fig.add_subplot(gs[0, 1])
df['my_start_date'] = pd.to_datetime(df['my_start_date'], errors='coerce')
start_date_counts = df[df['my_start_date'].notna()]['my_start_date'].dt.to_period('M').value_counts().sort_index()
start_date_counts.index = start_date_counts.index.to_timestamp()
start_date_counts.plot(kind='line', marker='o', ax=ax2, color='#FFD700', linewidth=2, markersize=8)
ax2.set_title('Watching Progress Over Time', fontsize=15, pad=15, fontweight='bold')

# Plot 3: Watched Episodes vs. Total Episodes
ax3 = fig.add_subplot(gs[0, 2])
sns.scatterplot(data=df, x='series_episodes', y='my_watched_episodes', hue='my_status', palette=anime_colors[:df['my_status'].nunique()], ax=ax3, s=80, edgecolor='black')
sns.lineplot(data=df, x='series_episodes', y='my_watched_episodes', color='#DC143C', ax=ax3, label='Trend', linewidth=2)
ax3.set_title('Watched Episodes vs. Total Episodes', fontsize=15, pad=15, fontweight='bold')
# Style legend for dark mode
legend = ax3.legend(loc='upper left', bbox_to_anchor=(1.02, 1), facecolor='#1E1E2E', edgecolor='#555566', labelcolor='white')

# ---------------------------------------------------------
# Plot 4: Top 10 Anime Series as Table
ax4 = fig.add_subplot(gs[1, 0])
top_10_anime = df.nlargest(10, 'my_score')[['series_title', 'my_score', 'series_type']].copy()
top_10_anime['series_title'] = top_10_anime['series_title'].apply(lambda x: textwrap.fill(str(x), width=25))

ax4.axis('off') 
ax4.set_title('Top 10 Rated Anime Series', fontsize=15, pad=15, fontweight='bold')
table = ax4.table(cellText=top_10_anime.values, colLabels=top_10_anime.columns, loc='center', cellLoc='center', bbox=[0, 0, 1, 1])
table.auto_set_font_size(False)
table.set_fontsize(10)
# Dark mode table styling
for (row, col), cell in table.get_celld().items():
    cell.set_edgecolor('#555566')
    if row == 0:
        cell.set_facecolor('#8A2BE2') # Purple header
        cell.set_text_props(color='white', weight='bold')
    else:
        cell.set_facecolor('#1E1E2E') # Dark rows
        cell.set_text_props(color='white')

# Plot 5: Distribution of Anime Watching Status
ax5 = fig.add_subplot(gs[1, 1])
status_counts = df['my_status'].value_counts()
_, texts, autotexts = ax5.pie(
    status_counts, 
    labels=status_counts.index, 
    autopct='%1.1f%%',
    colors=anime_cmap,
    wedgeprops=dict(width=0.6, edgecolor='#121212', linewidth=2),
    pctdistance=0.75,     
    labeldistance=1.15    
)
ax5.set_title('Distribution of Anime Watching Status', fontsize=15, pad=15, fontweight='bold')
# Ensure pie chart text is white to match dark theme
for text in texts + autotexts:
    text.set_color('white')
    text.set_fontsize(13)
    text.set_fontweight('bold')
    

# Plot 6: Completion Rate by Series Type
ax6 = fig.add_subplot(gs[1, 2])
df['completion_status'] = df.apply(lambda row: 1 if row['my_status'] == 'Completed' and row['my_watched_episodes'] == row['series_episodes'] else 0, axis=1)
completion_rate = df.groupby('series_type')['completion_status'].mean()
sns.barplot(x=completion_rate.index, y=completion_rate.values, palette=anime_colors, ax=ax6, edgecolor='#121212')
ax6.set_title('Completion Rate by Series Type', fontsize=15, pad=15, fontweight='bold')
ax6.tick_params(axis='x', rotation=45)

# ---------------------------------------------------------
# Plot 7: Anime Series Types Note Section
ax7 = fig.add_subplot(gs[2, 0])
note_text = """
Notes on Anime Series Types:
- TV: Regular television series
- Movie: Standalone films
- OVA: Original Video Animation
- Special: Special episodes or shorts
- ONA: Original Net Animation
"""
ax7.text(0.05, 0.5, note_text, fontsize=13, ha='left', va='center', color='white',
         bbox=dict(facecolor='#1E1E2E', edgecolor='#005A9C', alpha=0.9, boxstyle='round,pad=1', linewidth=2))
ax7.set_title('Anime Series Types Notes', fontsize=15, pad=15, fontweight='bold')
ax7.axis('off') 

# Plot 8: Score Distribution by Series Type
ax8 = fig.add_subplot(gs[2, 1])
sns.boxplot(data=df, x='series_type', y='my_score', palette=anime_colors, ax=ax8, showfliers=False, 
            boxprops=dict(edgecolor='#121212', linewidth=1.5), medianprops=dict(color='white', linewidth=2))
ax8.set_title('Score Distribution by Series Type', fontsize=15, pad=15, fontweight='bold')
ax8.tick_params(axis='x', rotation=45)

# Plot 9: Episodes Watched vs. Score
ax9 = fig.add_subplot(gs[2, 2])
sns.scatterplot(data=df, x='my_watched_episodes', y='my_score', hue='series_type', palette=anime_colors[:df['series_type'].nunique()], ax=ax9, s=70, edgecolor='#121212', alpha=0.9)
sns.regplot(data=df, x='my_watched_episodes', y='my_score', scatter=False, color='#00FFFF', ax=ax9, line_kws={'linewidth':2})
ax9.set_title('Episodes Watched vs. Score', fontsize=15, pad=15, fontweight='bold')
ax9.legend(facecolor='#1E1E2E', edgecolor='#555566', labelcolor='white')

# ---------------------------------------------------------
# Add bold frames/borders to all standard plots to make them pop against the dark background
plots_to_frame = [ax1, ax2, ax3, ax5, ax6, ax8, ax9]
for ax in plots_to_frame:
    for spine in ax.spines.values():
        spine.set_visible(True)
        spine.set_color('#8A2BE2') # Shadow Purple borders
        spine.set_linewidth(2) 

# Adjust layout, reserving top 10% for Main Title and KPIs
# Adjust layout, reserving top 14% (instead of 10%) for Main Title and KPIs
plt.tight_layout(rect=[0, 0, 1, 0.86])
plt.show()
```
