```python
print("Data Quality Summary: \n")

# Null value:
null_count = df.isna().sum()

# Empty-like values (empty string or false string)
empty_like = ["na", "n/a", "nan", "none", "unknown", "-", ""]
empty_like_count = df.apply(lambda col: col.astype(str).str.strip().str.lower().isin(empty_like).sum())

# Combine together
summary = pd.DataFrame({
    "null count": null_count,
    "Empty value": empty_like_count
}).fillna(0)

# Total Missing:
summary["Total Missing"] = summary["null count"] + summary["Empty value"]

# Percentage
summary["Missing %"] = (summary["Total Missing"] / len(df) * 100)

# Problematic rows and columns
summary = summary[summary["Total Missing"] > 0]

# Unique value
summary["Unique value"] = df.nunique()

# Data type
summary["Data Type"] = df.dtypes

# Duplicate value
dup_val = df.duplicated().sum()

# Sort
summary = summary.sort_values(by="Total Missing", ascending=False)

print(summary)
print(f"Duplicate Values: {dup_val}")
```

## Data cleaning

```python
# Define the anomaly placeholders we found
placeholders = [-1, '-1', 'na', 'Unknown / Non-Applicable', '']
total_rows = len(df)

# Calculate percentage of errors/missing values BEFORE cleaning
before_missing_pct = ((df.isin(placeholders).sum() + df.isna().sum()) / total_rows) * 100
```

```python
# Handle Missing values 

# Replace variations of missing data with np.nan
missing_indicators = [-1, '-1', 'na', 'Unknown / Non-Applicable']
df.replace(missing_indicators, np.nan, inplace=True)
```

```python
# droping unnecessary coloums 
# 1. List of extra feature columns to drop
columns_to_drop = [
    'my_comments', 'my_tags', 'my_sns', 'my_discuss', 
    'my_rewatch_value', 'my_rewatching', 'my_rewatching_ep', 
    'update_on_import', 'my_priority', 'my_times_watched', 
    'my_storage','my_storage_value','my_id', 'my_rated'
]

# 2. Automatically find all columns that start with "user_"
user_columns = [col for col in df.columns if col.startswith('user_')]

# 3. Combine both lists and drop them 
# (errors='ignore' prevents errors if a column is already gone)
df.drop(columns=columns_to_drop + user_columns, inplace=True, errors='ignore')

# 4. CRITICAL: Drop the very first row (Index 0) because it only contained user stats
df = df.iloc[1:].reset_index(drop=True)

# Look at your beautifully clean dataset!
display(df.head())
```

```python
# Save Cleand dataset
df.info()
df.isnull().sum()

# save to csv file 
df.to_csv("myanimelist_cleaned.csv", index=False)
```

## Data Quality graph after cleaning

```python
# Calculate percentage of missing values AFTER cleaning
after_missing_pct = (df.isna().sum() / total_rows) * 100

# Combine into a single DataFrame for easy plotting
comparison_df = pd.DataFrame({
    'Before Cleaning (%)': before_missing_pct,
    'After Cleaning (%)': after_missing_pct
})

# Keep only columns that actually had errors so the chart isn't cluttered
comparison_df = comparison_df[comparison_df['Before Cleaning (%)'] > 0].sort_values(by='Before Cleaning (%)', ascending=False)

# Plot the Before vs After comparison
plt.figure(figsize=(12, 6))
comparison_df.plot(kind='bar', figsize=(14, 6), color=['#d9534f', '#5cb85c'], width=0.8)

plt.title('Data Quality Check: Anomalies & Missing Values (Before vs. After Cleaning)', fontsize=14)
plt.ylabel('Percentage of Data (%)', fontsize=12)
plt.xlabel('Dataset Columns', fontsize=12)
plt.xticks(rotation=45, ha='right')
plt.legend(loc='upper right')
plt.tight_layout()
plt.show()
```
